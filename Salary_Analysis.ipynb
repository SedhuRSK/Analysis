```python
path = '/content/drive/MyDrive/salary_data_cleaned.csv'
import pandas as pd
df = pd.read_csv(path)
```

```python
columns_to_drop = ['Job Description', 'Competitors', 'company_txt']

data_cleaned = df.drop(columns=columns_to_drop)

data_cleaned['min_salary'] = pd.to_numeric(data_cleaned['min_salary'], errors='coerce')
data_cleaned['max_salary'] = pd.to_numeric(data_cleaned['max_salary'], errors='coerce')
data_cleaned['avg_salary'] = pd.to_numeric(data_cleaned['avg_salary'], errors='coerce')
```

```python
import matplotlib.pyplot as plt

job_counts = data_cleaned['Job Title'].value_counts().head(10)
job_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Job Titles by Count')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
job_counts = data_cleaned['Job Title'].value_counts()
job_counts = data_cleaned['Job Title'].value_counts()

print(job_counts)
job_counts.to_csv('job_counts.csv', header=['Count'], index_label='Job Title')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data_cleaned['avg_salary'], bins=20, kde=True, color='purple')
plt.title('Average Salary Distribution')
plt.xlabel('Average Salary')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
job_title_counts = data_cleaned['Job Title'].value_counts()
print(job_title_counts)

company_name_counts = data_cleaned['Company Name'].value_counts()
print(company_name_counts)
```

```python
total_rows = len(data_cleaned)
print(f"Total number of rows: {total_rows}")
```

```python

avg_salary_count = data_cleaned['avg_salary'].count()
print(f"Number of non-missing average salaries: {avg_salary_count}")
```

```python
# Count the number of jobs with an average salary greater than 100000
high_salary_count = (data_cleaned['avg_salary'] > 100000).sum()
print(f"Number of jobs with average salary > 100000: {high_salary_count}")
```

```python

location_counts = data_cleaned['Location'].value_counts().head(10)

print("Top 10 Locations by Job Count:")
print(location_counts)

location_counts.plot(kind='bar', color='blue')
plt.title('Top 10 Locations by Job Count')
plt.xlabel('Location')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45)
plt.show()
```

```python
skills = ['python_yn', 'R_yn', 'spark', 'aws', 'excel']

skill_demand = data_cleaned[skills].sum()

print("Skill Demand Count:")
print(skill_demand)

skill_demand.sort_values(ascending=False).plot(kind='bar', color='green')
plt.title('Demand for Skills in Job Market')
plt.xlabel('Skills')
plt.ylabel('Number of Jobs Requiring Skill')
plt.xticks(rotation=0)
plt.show()
```
